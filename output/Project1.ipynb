# Question #1 : Does your girlfriend value your achivement?

At the age of 17, when I was a high school kid, there was a beautiful girl in my class who I was obsessed with. I tried hard to get higher grade to draw her attention. However, she got in relationship with the little rascal with yellow hair in my class. I was really confused and kept asking myself the same question everyday: where did I lose? Several years later, I finally find the answer and the results are shown below.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('cleaned_hm.csv')

# To display the first few rows of the DataFrame
#df.head()
```

```python
# Filtering entries that contain the word 'boyfriend' in the 'cleaned_hm' column
df_with_boyfriend = df[df['cleaned_hm'].str.contains('boyfriend', case=False, na=False)]

# Displaying the first few rows of the filtered DataFrame
#df_with_boyfriend.head()
```

```python
# Calculating the frequency and percentage of each category in the 'predicted_category' column for the filtered data
predicted_category_filtered_frequency = df_with_boyfriend['predicted_category'].value_counts()
predicted_category_filtered_percentage = (predicted_category_filtered_frequency / predicted_category_filtered_frequency.sum()) * 100

# Creating the pie chart for predicted category distribution in the filtered data
plt.figure(figsize=(8, 8))
predicted_category_filtered_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title('Predicted Category Distribution in Filtered Data (Percentage)')
plt.ylabel('')  # Hiding the y-label for aesthetics
plt.show()
```

From the chart above, 96% happiness comes from the affection provided by boyfriends while achievement only takes 2%. It is obvious that it is not the achievement that makes a girl happy but affection, which remind me the days that the little rascal took my favorite girl for a ride along the shore, enjoying the breeze from the sea. Every their post on social media I stalked still appears in my mind nowadays. 

This result also gives insghts on how to date with a girl you like: do not talk too much on what achievement you've made or how successful you are on your career path because girls do not care about it, moreover, they may see this as distasteful showing off. Instead, talk more about fun stories happened on you. It could be an embarrasing experience when you were a kid, some fantastic places you have been to, or interesting people you've met.

# Question #2: What kind of activity does your girlfriend like the most?

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.tokenize import word_tokenize

def clean_and_split_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text.lower().split()

from collections import Counter
```

```python
# Extract and analyze words
all_words_filtered = df_with_boyfriend['cleaned_hm'].apply(clean_and_split_text).sum()
filtered_words_filtered = [word for word in all_words_filtered if word not in stopwords.words('english') and word != 'boyfriend']
tagged_words_filtered = pos_tag(filtered_words_filtered)
nouns_filtered = [word for word, tag in tagged_words_filtered if tag.startswith('NN')]
top_10_nouns_filtered = Counter(nouns_filtered).most_common(10)

#print(top_10_nouns_filtered)
```
